```python
# Student Performance Analytics System

students = {}

# -------------------- Input --------------------
n = int(input("Enter number of students: "))

for i in range(n):
    print(f"\nEnter details of Student {i+1}")

    name = input("Name: ")
    roll = input("Roll Number: ")
    branch = input("Branch: ")

    marks = []
    for j in range(5):
        mark = float(input(f"Enter marks of Subject {j+1}: "))
        marks.append(mark)

    total = sum(marks)
    percentage = total / 5

    if percentage >= 90:
        grade = "A+"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    elif percentage >= 50:
        grade = "D"
    else:
        grade = "Fail"

    students[roll] = {
        "Name": name,
        "Branch": branch,
        "Marks": marks,
        "Total": total,
        "Percentage": percentage,
        "Grade": grade
    }

# -------------------- Report Cards --------------------
print("\n========== STUDENT REPORT CARDS ==========")

for roll, data in students.items():
    print("\n--------------------------------")
    print("Roll Number :", roll)
    print("Name        :", data["Name"])
    print("Branch      :", data["Branch"])
    print("Marks       :", data["Marks"])
    print("Total       :", data["Total"])
    print("Percentage  : {:.2f}%".format(data["Percentage"]))
    print("Grade       :", data["Grade"])

# -------------------- Highest & Lowest --------------------
highest = max(students.items(), key=lambda x: x[1]["Percentage"])
lowest = min(students.items(), key=lambda x: x[1]["Percentage"])

print("\n========== CLASS ANALYTICS ==========")
print("Highest Scorer :", highest[1]["Name"], "-", highest[1]["Percentage"])
print("Lowest Scorer  :", lowest[1]["Name"], "-", lowest[1]["Percentage"])

# -------------------- Class Average --------------------
avg = sum(student["Percentage"] for student in students.values()) / len(students)
print("Class Average : {:.2f}%".format(avg))

# -------------------- Pass & Fail --------------------
passed = sum(1 for student in students.values() if student["Grade"] != "Fail")
failed = len(students) - passed

print("Students Passed :", passed)
print("Students Failed :", failed)

# -------------------- Subject-wise Topper --------------------
print("\n========== SUBJECT-WISE TOPPERS ==========")

for subject in range(5):
    topper = max(students.items(), key=lambda x: x[1]["Marks"][subject])
    print(f"Subject {subject+1}: {topper[1]['Name']} ({topper[1]['Marks'][subject]})")

# -------------------- Search by Roll Number --------------------
roll_search = input("\nEnter Roll Number to Search: ")

if roll_search in students:
    s = students[roll_search]
    print("\nStudent Found")
    print("Name:", s["Name"])
    print("Branch:", s["Branch"])
    print("Marks:", s["Marks"])
    print("Percentage: {:.2f}%".format(s["Percentage"]))
    print("Grade:", s["Grade"])
else:
    print("Student not found.")

# -------------------- Sort by Percentage --------------------
print("\n========== STUDENTS SORTED BY PERCENTAGE ==========")

sorted_students = sorted(
    students.items(),
    key=lambda x: x[1]["Percentage"],
    reverse=True
)

for roll, data in sorted_students:
    print(f"{data['Name']} ({roll}) - {data['Percentage']:.2f}%")

# -------------------- Remove Duplicate Branches --------------------
branches = set()

for student in students.values():
    branches.add(student["Branch"])

print("\nUnique Branches:")
print(branches)
```
